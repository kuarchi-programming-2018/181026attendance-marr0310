## matplotlibを使ってグラフをプロット

まずはデータを読み込み，それをX, Yのデータにして返す関数を作りましょう。
csv とはcomma separated valuesです。テキストで開いてみればわかります。エクセルでも開けます。

```python
import numpy as np
x=np.arange(-3,3,0.1)
y=np.sin(x)
import matplotlib.pyplot as plt
% matplotlib inline
plt.title("title")
plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.plot(x,y)
plt.show()
```

matplotlibを用いてデータを散布図にプロットする関数を作成します。

```python
import matplotlib.pyplot as plt
% matplotlib inline
#  % matplotlib inlineはjupyter notebookでmatplotlibを行内に表示するためのおまじない
```

```python
def draw_graph(X, Y, title, xlabel, ylabel):
    # ここに描画するプログラムを書く。
    plt.title("title")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(X,Y)
    plt.show()
    # グラフタイトル
    # X軸タイトル
    # Y軸タイトル
    # グラフをメモリ上に作成
    plt.show()  # グラフの描画
```

```python
import csv
def read_data(file_name):
    reader = csv.reader(open(file_name, 'r'))
    X = []
    Y = []
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
    return X, Y  # データを返す
```

では，データを読み込んで描画して見ましょう。

```python
import csv
def read_data(file_name):
    reader = csv.reader(open(file_name, 'r'))
    X = []
    Y = []
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
    return X, Y  # データを返す

file_name = 'data3.csv'  # ファイル名
x,y=read_data(file_name)
print(x)
print(y)
# 関数read_dataによりデータを取得
# タイトル，ラベル
title="data from csv"
xl="x values"
yl="y values"

draw_graph(x, y, title, xl, yl) # 関数draw_graphによりグラフ描画
```
